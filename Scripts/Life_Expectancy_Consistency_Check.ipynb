```python
# Define the range of years you want to impute (2000 to 2015)
years_to_impute = range(2000, 2016)

# Iterate through each row in the primary dataset
for index, row in df1.iterrows():
    country_name = row['Country']  # Replace with the actual column name for country names
    year = row['Year']  # Replace with the actual column name for years
    
    # Check if the year is within the range to impute (2000 to 2015)
    if year in years_to_impute:
        # Find the corresponding row in the second dataset
        second_dataset_row = df2[(df2['Country Name'] == country_name) & (df2['Indicator Name'] == 'GDP per capita (current US$)')]
        
        # Check if the second dataset row exists for the given country
        if not second_dataset_row.empty:
            # Extract GDP and population data from the second dataset
            gdp_value = second_dataset_row[str(year)].values[0]
            
            # Update the GDP and population columns in the primary dataset
            df1.at[index, 'GDP'] = gdp_value  # Replace with the actual column name for GDP
            
# Save the updated primary dataset with imputed values
output_file_path = "path_to_output_dataset.csv"
df1.to_csv('C:\\Users\\Ankit\\Downloads\\output_dataset.csv', index=False)
```

```python
import numpy as np

# Define a function to impute missing population values
def impute_population(row):
    country = row['Country']
    year = row['Year']
    population = df2.loc[df2['Country Name'] == country, str(year)].values
    if len(population) > 0 and not np.isnan(population[0]):
        return population[0]
    else:
        return np.nan

# Apply the impute_population function to fill missing population values
df1['Population'] = df1.apply(impute_population, axis=1)

# Save the updated primary dataset with imputed population values
df1.to_csv(r'C:\Users\Ankit\Downloads\output_dataset.csv', index=False)
```

```python
import pandas as pd

# Load the cleaned dataset
cleaned_data = pd.read_csv('C:/Users/Ankit/OneDrive/Documents/CareerFoundry/Final Projects/WHO MORTALITY/Data/Cleaned Data/cleaned_life_expectancy_data.csv')

# Load the complete population dataset
population_data = pd.read_csv('C:/Users/Ankit/Downloads/output_dataset.csv')

# Rename columns in the population dataset to match cleaned_data
population_data.rename(columns={
    'Life expectancy ': 'Life expectancy',
    'Measles ': 'Measles',
    ' BMI ': 'BMI',
    'Diphtheria ': 'Diphtheria',
    ' HIV/AIDS': 'HIV/AIDS',
    ' thinness  1-19 years': 'thinness 1-19 years',
    ' thinness 5-9 years': 'thinness 5-9 years'
}, inplace=True)

# Merge the datasets and handle overlapping columns with suffixes
merged_data = pd.merge(cleaned_data, population_data, on=['Country', 'Year'], how='left', suffixes=('_x', '_y'))

# Fill missing values in Population_x and Population_y with non-missing values
merged_data['Population_x'].fillna(merged_data['Population_y'], inplace=True)
merged_data['Population_y'].fillna(merged_data['Population_x'], inplace=True)

# Remove extra spaces from column names
merged_data.columns = merged_data.columns.str.strip()

# Get the list of columns available in the merged dataset
available_columns = merged_data.columns.tolist()

# Manually specify the columns to save based on the available columns
columns_to_save = [
    'Country', 'Year', 'Status', 'Life expectancy', 'Adult Mortality', 'infant deaths',
    'Alcohol', 'percentage expenditure', 'Hepatitis B', 'Measles', 'BMI', 'under-five deaths',
    'Polio', 'Total expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP', 'Population_x', 'Population_y',
    'thinness 1-19 years', 'thinness 5-9 years', 'Income composition of resources', 'Schooling'
]

# Filter the columns to save based on the available columns
columns_to_save = [col for col in columns_to_save if col in available_columns]

# Save the merged dataset with imputed population values and the specified columns
merged_data = merged_data[columns_to_save]

# Save the merged dataset
merged_data.to_csv('C:/Users/Ankit/OneDrive/Documents/CareerFoundry/Final Projects/WHO MORTALITY/Data/Cleaned Data/cleaned_life_expectancy_data_with_population.csv', index=False)
```
